```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
berka = pd.read_csv("C:/Users/niafrate/OneDrive - Ryerson University/CIND820/finData.csv")
```

```python
berka.info()
```

```python
berka = berka.drop(labels='account_id', axis =1)
```

```python
enc_stmFreq = (berka.groupby('frequency').size())/len(berka)
enc_stmFreq
```

```python
enc_cardType = (berka.groupby('card_type').size())/len(berka)
enc_cardType
```

```python
berka['frequency'] = berka['frequency'].apply(lambda x : enc_stmFreq[x])
berka['card_type'] = berka['card_type'].apply(lambda x : enc_cardType[x])
```

```python
berka.head(5)
```

```python
berka.info()
```

```python
#One hot encoding
dum1 = pd.get_dummies(berka['gender'])
dum2 = pd.get_dummies(berka['same_district'])
```

```python
berka = berka.join(dum1)
```

```python
berka = berka.join(dum1)
berka.rename(columns={0:'Gender_F'}, inplace = True)
berka.rename(columns={1:'Gender_M'}, inplace = True)
```

```python
berka.head()
```

```python
berka = berka.drop(labels='gender', axis =1)
```

```python
berka = berka.join(dum2)
```

```python
berka.rename(columns={0:'same_dist_N'}, inplace=True)
berka.rename(columns={1:'same_dist_Y'}, inplace=True)
```

```python
berka=berka.drop(labels='same_district', axis = 1)
```

```python
berka.head(5)
```

```python
berka.shape
```

```python
colnames = list(berka.columns.values)
print(colnames)
```

```python
berka = berka[['amount', 'duration', 'payments', 'frequency', 'date_diff', 'card_type', 'no_orders', 'avg_order', 'avg_trans', 'avg_bal', 'no_of_trans', 'total_inhabitants', 'avg_unemp_rt', 'avg_crime_rt', 'avg_salary', 'age_at_loan', 'Gender_F', 'Gender_M', 'same_dist_N', 'same_dist_Y', 'loan_status']]
```

```python
berka.head(5)
```

```python
fig, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(berka.corr(), cmap = sns.diverging_palette(220, 10, as_cmap = True), 
            square=True, cbar=False, ax=ax, annot=True, linewidths=0.1,vmax=1.0, 
            linecolor='white', annot_kws={'fontsize':12 })
plt.show()
```
